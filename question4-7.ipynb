Code Implementation Task: Implement code to fill the missing data (impute) in daily_vaccinations column per country with the minimum daily vaccination number of relevant countries.  
Note: If a country does not have any valid vaccination number yet, fill it with “0” (zero). 



```python
import pandas as pd

# Read data from CSV file
df = pd.read_csv('/Users/emirysaglam/Downloads/piworks/country_vaccination_stats.csv')
df

```

```python
# Calculating the minimum daily vaccination number for each country
min_daily_vaccinations = df.groupby('country')['daily_vaccinations'].min().reset_index()

min_daily_vaccinations
```

```python
# Merging the original dataframe with the minimum values
df = df.merge(min_daily_vaccinations, on='country', suffixes=('', '_min'))

# Filling missing values in daily_vaccinations with the minimum value
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(df['daily_vaccinations_min'])

df
```

```python
# Filling missing values with 0
df['daily_vaccinations'] = df['daily_vaccinations'].fillna(0)
df
```

```python
# Dropping the additional column
df.drop('daily_vaccinations_min', axis=1, inplace=True)

print(df)
```

#### **Question 6:**

Code Implementation Task: Implement code to list the top-3 countries with highest median daily vaccination numbers by considering missing values imputed version of dataset.

```python
# Calculating median daily vaccination number for each country
median_daily_vaccinations = df.groupby('country')['daily_vaccinations'].median().reset_index()

# Sorting by median daily vaccinations in descending order
median_daily_vaccinations_sorted = median_daily_vaccinations.sort_values(by='daily_vaccinations', ascending=False)

# Listing top-3 countries with highest median daily vaccination numbers
top_3_countries = median_daily_vaccinations_sorted.head(3)

print(top_3_countries)
```

#### **Question 7**

What is the number of total vaccinations done on 1/6/2021 (MM/DD/YYYY) by considering missing values imputed version of dataset?

```python
date_filter = df['date'] == '1/6/2021'  # Adjust date format as per your data
vaccinations_on_01062021 = df[date_filter]

# Calculate total vaccinations for 1/6/2021
total_vaccinations_01062021 = vaccinations_on_01062021['daily_vaccinations'].sum()

print("Total vaccinations done on 1/6/2021:", total_vaccinations_01062021)
```
